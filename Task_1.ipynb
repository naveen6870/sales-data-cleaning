```python
# Drop columns with too many missing values (e.g., ADDRESSLINE2)
df = df.drop(columns=['ADDRESSLINE2'])
```

```python
# Fill missing values (example: fill STATE with 'Unknown')
df['STATE'] = df['STATE'].fillna('Unknown')
```

```python
# 2. Remove duplicate rows
df = df.drop_duplicates()
```

```python
# 3. Standardize text values (e.g., COUNTRY names to lowercase)
df['COUNTRY'] = df['COUNTRY'].str.strip().str.lower()
```

```python
# 4. Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
```

```python
# Convert ORDERDATE to datetime and format to dd-mm-yyyy
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
df['ORDERDATE'] = df['ORDERDATE'].dt.strftime('%d-%m-%Y')
```

```python
# Preview the formatted date values
df['ORDERDATE'].head()
```

```python
# 5. Rename columns to lowercase and remove spaces
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# 6. Check and fix data types
# Convert 'postalcode' to string (in case of leading zeros)
df['postalcode'] = df['postalcode'].astype(str)
```

```python
# Confirm changes
print(df.dtypes.head())
```

```python
df.to_csv('cleaned_sales_data.csv', index=False)
```
